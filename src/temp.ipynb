```python
import numpy as np
import time
```

```python
data = {}
for n in range(1, 100, 5):
    ues = np.arange(n).astype(float)
    k = 0
    start = time.time()
    while True:
        flag = True
        for i in range(1, len(ues)):
            flag = flag and abs(ues[0]-ues[i])<1e-6
        if flag:
            break
        else:
            ues_ = ues.copy()
            for i in range(1, len(ues)):
                ues_[i] = (ues[i-1]+ues[i])/2
            ues_[0] = (ues[0]+ues[-1])/2
            ues = ues_.copy()
    end = time.time()
    dur = end-start
    data.update({n: dur})
```

```python
from matplotlib import pyplot as plt
import pandas as pd

plt.plot(data.keys(), data.values())
plt.xlabel('UEs')
plt.ylabel('Time(s)')
plt.grid()
plt.show()
```
