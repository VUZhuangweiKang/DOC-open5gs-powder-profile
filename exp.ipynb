```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('exp_results.csv')
avg = data.groupby(by='UEs').mean()
ues = data['UEs'].unique()
box_data = [data[data['UEs']==x]['time'] for x in ues]
plt.figure(figsize=(10,6))
plt.boxplot(box_data)
plt.xlabel('# UEs', fontsize=24)
plt.ylabel('Decision-reaching Time (s)', fontsize=24)
plt.grid()
plt.xticks(np.arange(1, 11), np.arange(3, 33, 3), fontsize=20)
plt.yticks(fontsize=20)
plt.savefig('results.png', dpi=300)
plt.show()
```
